```python
# Import pandas for data processing
import pandas as pd
```

```python
# Load the raw dataset created in extract step
df = pd.read_csv("raw_data.csv")

# View first 5 rows
df.head()
```

```python
# Convert views and likes to integer (numbers)
df['views'] = df['views'].astype(int)
df['likes'] = df['likes'].astype(int)

# Check data types
df.dtypes
```

```python
# Convert publish_date from text to datetime format
df['publish_date'] = pd.to_datetime(df['publish_date'])

# Check
df.head()
```

```python
# Create mapping for category IDs to names
category_map = {
    1: "Film & Animation",
    2: "Autos & Vehicles",
    10: "Music",
    15: "Pets & Animals",
    17: "Sports",
    20: "Gaming",
    22: "People & Blogs",
    23: "Comedy",
    24: "Entertainment",
    25: "News & Politics",
    26: "Howto & Style",
    27: "Education",
    28: "Science & Technology"
}
```

```python
# Convert category_id to category name
df['category'] = df['category_id'].astype(int).map(category_map)

# Check result
df[['category_id', 'category']].head()
```

```python
# Create engagement rate column
# (likes divided by views)
df['engagement_rate'] = df['likes'] / df['views']

# View sample
df[['views', 'likes', 'engagement_rate']].head()
```

```python
# Remove rows with missing values
df.dropna(inplace=True)

# Check if any null values remain
df.isnull().sum()
```

```python
# Remove category_id column (not needed anymore)
df.drop(columns=['category_id'], inplace=True)

# Check final structure
df.head()
```

```python
# Save cleaned dataset
df.to_csv("processed_data.csv", index=False)

print("✅ Processed data saved successfully")
```
