```python
# Import required libraries
from googleapiclient.discovery import build
import pandas as pd
```

```python
# Add your YouTube API key here
API_KEY = "AIzaSyAmUiKC_D_g-bGFvZXFJqxXu-SZ73rQIUk"
# Create connection to YouTube API
youtube = build('youtube', 'v3', developerKey=API_KEY)
```

```python
# Request trending videos from India
request = youtube.videos().list(
    part="snippet,statistics",   # video details + performance
    chart="mostPopular",         # trending videos
    regionCode="IN",             # India region
    maxResults=50                # number of videos
)
```

```python
# Send request to YouTube and get response
response = request.execute()
```

```python
# Create empty list to store extracted data
data = []
# Loop through each video and extract required fields
for item in response['items']:
    data.append({
        'title': item['snippet']['title'],                # Video title
        'channel': item['snippet']['channelTitle'],       # Channel name
        'views': item['statistics'].get('viewCount', 0),  # Views (default 0)
        'likes': item['statistics'].get('likeCount', 0),  # Likes (default 0)
        'category_id': item['snippet']['categoryId'],     # Category ID
        'publish_date': item['snippet']['publishedAt']    # Upload date
    })
```

```python
# Convert list into table format (DataFrame)
df = pd.DataFrame(data)

# Show first 5 rows
df.head()
```

```python
# Save data to CSV file
df.to_csv("raw_data.csv", index=False)

print("✅ Data saved successfully")
```
