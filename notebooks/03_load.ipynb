```python
# Import required libraries
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Load the cleaned dataset from transform step
df = pd.read_csv("processed_data.csv")

# Preview data
df.head()
```

```python
# Create connection to MySQL database
# Format: mysql+pymysql://username:password@host:port/database_name

engine = create_engine("mysql+pymysql://root:1234@127.0.0.1:3306/youtube_db")
```

```python
# Upload data to MySQL table
df.to_sql(
    name="trending_videos",   # table name
    con=engine,              # connection
    if_exists='replace',     # replace if table already exists
    index=False              # don't add index column
)

print("✅ Data loaded into MySQL successfully")
```
